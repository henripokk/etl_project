## vaatan kas saab andmed paremini esitatud

```python
def data(slider_value, multi_choice_value):
    df_kaart=df.loc[pd.to_datetime(slider_value[0]):pd.to_datetime(slider_value[1])]
    df_kaart = df_kaart.loc[df_kaart[multi_choice_value].sum(axis=1)>0 ]
    return df_kaart

data(slider.value, multi_choice.value)[multi_choice.value]
```

```python
# suured numbrid - hukkunud, vigastatud ja nnetusi
@pn.depends( pn.widgets.DateRangeSlider.param.value_throttled)
@pn.depends( pn.widgets.MultiChoice.param.value)
def statistika(slider_value, multi_choice_value, veerg):
    if veerg=="nnetusi":
        return pn.indicators.Number(name=veerg, value=data(slider_value, multi_choice_value)['Juhtumi nr'].count(), font_size = "36pt", width=150)  
    else:
        return pn.indicators.Number(name=veerg, value=data(slider_value, multi_choice_value)[veerg].sum(), font_size = "36pt", width=150)

pn.Row(statistika(slider.value, multi_choice.value, 'nnetusi'),statistika(slider.value, multi_choice.value, 'nnetusi'))


```

```python
import plotly.graph_objects as go

@pn.depends(pn.widgets.RadioButtonGroup.param.value)   
def osalejad(liik, multi_choice_value, slider_value):
    df_kaart = data(slider_value, multi_choice_value)  
    options = ['Jalakija osalusel', 'Kergliikurijuhi osalusel', 'Kaassitja osalusel', 'Maastikusiduki juhi osalusel', "Eaka (65+) mootorsidukijuhi osalusel",  "Bussijuhi osalusel", "Veoautojuhi osalusel", "hissidukijuhi osalusel", "Siduautojuhi osalusel", "Mootorratturi osalusel", "Mopeedijuhi osalusel", "Jalgratturi osalusel", "Alaealise osalusel", "Turvavarustust mitte kasutanud isiku osalusel", "Esmase juhiloa omaniku osalusel", "Mootorsidukijuhi osalusel"]

    # Count the number of accidents for each participant type
    df_count = df_kaart[options].apply(pd.Series.value_counts).T.reset_index().sort_values(1.0)
    # Counting the number of accidents for each time component
    nnetusi = df_kaart['Liiklusnnetuse liik [3]'].value_counts().sort_index().sort_values()

    # Create the plot using Plotly Express
    fig1 = go.Figure()
    fig1.add_trace(go.Bar(y=df_count['index'], x=df_count[1.0], orientation='h'))
    fig1.update_layout(
        title='nnetustes osalejate arv',
        xaxis_title='Kokku',
        yaxis_title='',
        xaxis=dict(showgrid=True, gridcolor='lightgrey'),  # Show x-axis gridlines with grey color
        yaxis=dict(showgrid=True, gridcolor='lightgrey'),  # Show y-axis gridlines with grey color
        plot_bgcolor='white'
    )

    # Creating bar plots using Plotly
    if liik == "Osalejad":
        return fig1
    elif liik == "nnetuse liik": 
        fig = go.Figure(go.Bar(y=nnetusi.index, x=nnetusi.values, orientation='h'))
        fig.update_layout(
            title='Erinevat liiki nnetuste arv',
            xaxis_title='nnetuste arv',
            yaxis_title='',
            xaxis=dict(showgrid=True, gridcolor='lightgrey'),  # Show x-axis gridlines with grey color
            yaxis=dict(showgrid=True, gridcolor='lightgrey'),  # Show y-axis gridlines with grey color
            plot_bgcolor='white'
        )
        return fig

osalejad(nnetused.value, multi_choice.value, slider.value)


```

```python
import plotly.express as px

@pn.depends(pn.widgets.DateRangeSlider.param.value_throttled)
def onnetused_joonis(slider_value, multi_choice_value):
    df_kaart = data(slider_value, multi_choice_value)
    
    # Counting the number of accidents for each time component
    nnetusi = df_kaart['Liiklusnnetuse liik [3]'].value_counts().sort_index().sort_values()

    # Creating bar plots using Plotly
    fig = px.bar(y=nnetusi.index, x=nnetusi.values, labels={'y': 'nnetuse liik', 'x': 'nnetuste arv'}, title='Erinevat liiki nnetuste arv')

    fig.update_layout(
        plot_bgcolor='rgba(0, 0, 0, 0)',  # Set background color to transparent
        paper_bgcolor='rgba(0, 0, 0, 0)',  # Set plot area background color to transparent
        xaxis=dict(showgrid=True, gridcolor='lightgray'),  # Show grid lines for x-axis
        yaxis=dict(showgrid=True, gridcolor='lightgray')  # Show grid lines for y-axis
    )
    
    return fig

onnetused_joonis(slider.value, multi_choice.value)

```

```python
import plotly.graph_objects as go

@pn.depends(pn.widgets.RadioButtonGroup.param.value)
def perioodid_joonis(perioodid, slider_value, multi_choice_value):
    df_kaart = data(slider_value, multi_choice_value)

    # Counting the number of accidents for each time component
    Aastad = df_kaart['year'].value_counts().sort_index()
    Kuud = df_kaart['month'].value_counts().sort_index()
    Pevad = df_kaart['day_of_week'].value_counts().sort_index()
    Tunnid = df_kaart['hour_of_day'].value_counts().sort_index()

    # Creating line plots using Plotly
    if perioodid == "Aastad":
        fig = go.Figure(data=go.Scatter(x=Aastad.index, y=Aastad.values, mode='lines+markers'))
    elif perioodid == "Kuud":
        fig = go.Figure(data=go.Scatter(x=Kuud.index, y=Kuud.values, mode='lines+markers'))
    elif perioodid == "Pevad":
        fig = go.Figure(data=go.Scatter(x=Pevad.index, y=Pevad.values, mode='lines+markers'))
    elif perioodid == "Tunnid":
        fig = go.Figure(data=go.Scatter(x=Tunnid.index, y=Tunnid.values, mode='lines+markers'))

    fig.update_layout(
        title='nnetusi perioodide kaupa',
        xaxis_title='Periood',
        yaxis_title='nnetuste arv',
        plot_bgcolor='rgba(0,0,0,0)',  # Set background color to transparent
        xaxis=dict(showgrid=True, gridcolor='lightgray'),  # Show grid lines for x-axis
        yaxis=dict(showgrid=True, gridcolor='lightgray')  # Show grid lines for y-axis
    )
    return fig

perioodid_joonis(perioodid.value, slider.value, multi_choice.value)

```
